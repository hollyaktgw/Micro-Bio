```python
# import data analysis libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing tsO45G Secretory Membrane (Labeled CFP) data
data = pd.read_csv('C:/Users/holly/OneDrive/Bureau/Internship/Processed Images/MyExpt_SecratoryMembrane.csv')
data.head()

# give me the names of the intensity columns
data.columns
```

```python
# select the mean intensity columns 'Intensity_MeanIntensity_tsO45G'
data_mean = data[['Intensity_MeanIntensity_tsO45G']]
data_mean.head()
```

```python
# make a histogram of the mean intensity
plt.hist(data_mean['Intensity_MeanIntensity_tsO45G'], bins=100)
plt.xlabel('Mean Intensity')
plt.ylabel('Frequency')
plt.title('Mean Intensity of tsO45G Secretory Membrane')
plt.show()
```

```python
# what is the mean of the mean intensity?
data_mean['Intensity_MeanIntensity_tsO45G'].mean()

# what is the average of the mean intensity?
data_mean['Intensity_MeanIntensity_tsO45G'].median()

```

```python
# do the same for the standard deviation
data_std = data[['Intensity_StdIntensity_tsO45G']]
plt.hist(data_std['Intensity_StdIntensity_tsO45G'], bins=100)
plt.xlabel('Standard Deviation')
plt.ylabel('Frequency')
plt.title('Standard Deviation of tsO45G Secretory Membrane')
plt.show()
```

```python
# compare center mass intensity and edge intensity
data_center = data[['Intensity_MeanIntensity_tsO45G_RegionCenterMassIntensity']]
data_edge = data[['Intensity_MeanIntensity_tsO45G_RegionEdgeIntensity']]
plt.scatter(data_center, data_edge)
plt.xlabel('Center Mass Intensity')
plt.ylabel('Edge Intensity')
plt.title('Center Mass Intensity vs Edge Intensity')
plt.show()
```
